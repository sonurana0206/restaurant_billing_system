```python
from tkinter import *
from tkinter import ttk
import random
from datetime import datetime
from tkinter import messagebox
import sys
operator ="" 
def main():
    win=Tk()
    app=LoginPage(win)
    win.mainloop()
class LoginPage:
    def __init__(self,win):
        self.win=win
        self.win.geometry("1350x750+0+0")
        self.win.title("Restaurant Billing System")
        self.win.config(bg="black")
        self.title_label=Label(self.win,text="RESTAURANT BILLING SYSTEM",font=('algerian',35,'bold'),bg="black",fg="orange",bd=8,relief='groove')
        self.title_label.pack(side=TOP,fill=X)
        self.main_frame=Frame(self.win,bg="lightgrey",bd='6',relief=GROOVE)
        self.main_frame.place(x=250,y=150,width=800,height=400)
        self.login_lbl=Label(self.main_frame,text="LOGIN",bg="black",fg='white',font=('aria',25,'bold'),bd=6,relief='groove',anchor=CENTER)
        self.login_lbl.pack(side=TOP,fill=X)
        self.entry_frame=LabelFrame(self.main_frame,text='Enter Details',bd=6,relief='groove',bg='lightgrey',font=('aria',18,'bold'))
        self.entry_frame.pack(fill=BOTH,expand=True)
        self.enter_lbl=Label(self.entry_frame,text="Enter Username  ",bg="lightgrey",font=('aria',20,'bold'))
        self.enter_lbl.grid(row=0,column=0,padx=2,pady=2)
        self.enter1_lbl=Label(self.entry_frame,text="Enter Password  ",bg="lightgrey",font=('aria',20,'bold'))
        self.enter1_lbl.grid(row=2,column=0,padx=2,pady=2)
        #-------------------Entry---------------------------------------
        username=StringVar()
        password=StringVar()
        self.enter_user=Entry(self.entry_frame,font=('aria',20),bd=6,textvariable=username)
        self.enter_user.grid(row=0,column=1,padx=2,pady=2)
        self.enter_pass=Entry(self.entry_frame,font=('aria',20),bd=6,textvariable=password,show="*")
        self.enter_pass.grid(row=2,column=1,padx=2,pady=2)
        #------------------Functions------------------------------------
        
        def login():
            if username.get()=="sonu" and password.get()=="1234":
                self.bill_btn.config(state="normal")
            else:
                pass
        def reset():
            username.set("")
            password.set("")
        def bill():
            self.root=Toplevel(self.win)
            self.app=Window2(self.root)
            
        
        
        #------------------Buttons--------------------------------------
        self.button_frame=LabelFrame(self.entry_frame,text="Options",font=('aria',15,'bold'),bg="lightgrey",bd=7,relief="groove")
        self.button_frame.place(x=20,y=150,width=730,height=85)
        self.login_btn=Button(self.button_frame,text='LOGIN',font=('aria',15,'bold'),bd=5,bg="white",fg='black',width=15,command=login)
        self.login_btn.grid(row=0,column=0,padx=20,pady=3)
        self.bill_btn=Button(self.button_frame,text='BILLING',font=('aria',15,'bold'),bd=5,bg="white",fg='black',width=15,command=bill)
        self.bill_btn.config(state="disabled")
        self.bill_btn.grid(row=0,column=1,padx=20,pady=3)
        self.re_btn=Button(self.button_frame,text='RESET',font=('aria',15,'bold'),bd=5,bg="white",fg='black',width=15,command=reset)
        self.re_btn.grid(row=0,column=2,padx=20,pady=3)
class Window2:
    def __init__(self,win):
        self.win=win
        self.win.geometry('1350x750+0+0')
        self.win.title("RBS")
        self.win.config(bg="black")

        self.title_label=Label(self.win,text="RESTAURANT BILLING SYSTEM",font=('algerian',35,'bold'),bg="black",fg="orange",bd=8,relief='groove')
        self.title_label.pack(side=TOP,fill=X)
        self.win.resizable(0,0)
        #-------------var----------------------------------------------------
        bill_no = random.randint(100,9999)
        bill_no_tk=IntVar()
        bill_no_tk.set(bill_no)
        calc_var=StringVar()
        cust_nm= StringVar()
        cust_cn=StringVar()
        dat=StringVar()
        item_pur=StringVar()
        item_qty = StringVar()
        cone=StringVar()
        dat.set(datetime.now())
        total_list=[]
        self.grd_total=0
        #---------------------Entry-------------------------------------------
        self.entry_frame=LabelFrame(self.win,text="Enter Details",bg="lightgrey",font=("aria",20,'bold'),bd=7,relief="groove")
        self.entry_frame.place(x=20,y=95,width=500,height=580)
        
        self.l1=Label(self.entry_frame,text="Bill Number ",font=('aria',15,'bold'),bg="lightgrey")
        self.l1.grid(row=0,column=0,padx=2,pady=2)
        self.e1=Entry(self.entry_frame,bd=5,textvariable=bill_no_tk,font=('aria',15,'bold'))
        self.e1.grid(row=0,column=1,padx=2,pady=2)
        self.e1.config(state="disabled")
        
        self.l2=Label(self.entry_frame,text="Customer Name ",font=('aria',15,'bold'),bg="lightgrey")
        self.l2.grid(row=1,column=0,padx=2,pady=2)
        self.e2=Entry(self.entry_frame,bd=5,textvariable=cust_nm,font=('aria',15,'bold'))
        self.e2.grid(row=1,column=1,padx=2,pady=2)

        self.l3=Label(self.entry_frame,text="Customer Contact ",font=('aria',15,'bold'),bg="lightgrey")
        self.l3.grid(row=2,column=0,padx=2,pady=2)
        self.e3=Entry(self.entry_frame,bd=5,textvariable=cust_cn,font=('aria',15,'bold'))
        self.e3.grid(row=2,column=1,padx=2,pady=2)
        
        self.l4=Label(self.entry_frame,text="Date ",font=('aria',15,'bold'),bg="lightgrey")
        self.l4.grid(row=3,column=0,padx=2,pady=2)
        self.e4=Entry(self.entry_frame,bd=5,textvariable=dat,font=('aria',15,'bold'))
        self.e4.grid(row=3,column=1,padx=2,pady=2)
        
        self.l5=Label(self.entry_frame,text="Items Purchased ",font=('aria',15,'bold'),bg="lightgrey")
        self.l5.grid(row=4,column=0,padx=2,pady=2)
        self.e5=Entry(self.entry_frame,bd=5,textvariable=item_pur,font=('aria',15,'bold'))
        self.e5.grid(row=4,column=1,padx=2,pady=2)
        
        self.l6=Label(self.entry_frame,text="Item Quantity ",font=('aria',15,'bold'),bg="lightgrey")
        self.l6.grid(row=5,column=0,padx=2,pady=2)
        self.e6=Entry(self.entry_frame,bd=5,textvariable=item_qty,font=('aria',15,'bold'))
        self.e6.grid(row=5,column=1,padx=2,pady=2)
        
        self.l7=Label(self.entry_frame,text="Cost of One ",font=('aria',15,'bold'),bg="lightgrey")
        self.l7.grid(row=6,column=0,padx=2,pady=2)
        self.e7=Entry(self.entry_frame,bd=5,textvariable=cone,font=('aria',15,'bold'))
        self.e7.grid(row=6,column=1,padx=2,pady=2)
        #------------------------func----------------------------------
        def default_bill():
            self.bill_txt.insert(END,"\t\t\t\t   Sarang Restaurant")
            self.bill_txt.insert(END,"\n\t\t\t7 Street, Near Railway Lines, Edinburg")
            self.bill_txt.insert(END,"\n\t\t\t\tContact - +919850289851")
            self.bill_txt.insert(END,"\n*************************************************************************************")
            self.bill_txt.insert(END,f"\nBill Number {bill_no_tk.get()}")
            
        def gen():
            if cust_nm.get()=="" or (cust_cn.get() =="" or len(cust_cn.get())!=10):
                messagebox.showerror("Error!",f"Please enter all field correctly.",parent=self.win)
            else:
                self.bill_txt.insert(END,f"\t\t\t\t\t\tDate : {dat.get()}")
                self.bill_txt.insert(END,f"\nCustomer Name : {cust_nm.get()}")
                self.bill_txt.insert(END,f"\nCustomer Contact : {cust_cn.get()}")
                self.bill_txt.insert(END,"\n*************************************************************************************")
                self.bill_txt.insert(END,"\nProduct Name\t\t\tQuantity\t\t\tPer Cost\t\t\tTotal")
                self.bill_txt.insert(END,"\n*************************************************************************************")
                self.b1.config(state="normal")
                self.b4.config(state="normal")
               
        
        def clr():
            cust_nm.set("")
            cust_cn.set("")
            item_pur.set("")
            item_qty.set("")
            cone.set("")
            
        def reset():
            total_list.clear()
            self.grd_total = 0
            self.b1.config(state="disabled")
            self.b4.config(state="disabled")
            self.b6.config(state="disabled")
            self.bill_txt.delete("1.0",END)
            default_bill()
            
        def add1():
            if item_pur.get()==""or item_qty.get()=="":
                messagebox.showerror("Error!",f"Please enter fields correctly.",parent=self.win)
            qty=int(item_qty.get())
            cones=int(cone.get())
            total =qty*cones
            total_list.append(total)
            self.bill_txt.insert(END,f"\n{item_pur.get()}\t\t\t{item_qty.get()}\t\t\tRs.{cone.get()}\t\t\tRs.{total}")
            
        def total():
            for item in total_list:
                self.grd_total = self.grd_total + item
            self.bill_txt.insert(END,"\n*************************************************************************************")    
            self.bill_txt.insert(END,f"\t\t\t\t\t\t\t\tGrand Total : {self.grd_total}")
            self.bill_txt.insert(END,"\n*************************************************************************************") 
            self.b6.config(state="normal")
        
        def save():
            user_choice=messagebox.askyesno("Confirm ?",f"Do you want to save the bill {bill_no_tk.get()} ?",parent=self.win)
            if user_choice > 0:
                self.bill_content = self.bill_txt.get("1.0",END)
                try:
                    con= open(f"{sys.path[0]}/bills/"+str(bill_no_tk.get())+".txt","w")
                except Exception as e:
                    messagebox.showerror("Error!",f"Error due to {e}",parent=self.win)
                con.write(self.bill_content)
                con.close()
                messagebox.showinfo("Success!",f"Bill {bill_no_tk.get()} has been saved successfully!",parent=self.win)
            else:
                return
        
        
        #----------Buttons--------------------------------------------
        self.button_frame=LabelFrame(self.entry_frame,text="Options",bg="lightgrey",font=("aria",15,'bold'),bd=5,relief="groove")
        self.button_frame.place(x=20,y=280,width=400,height=250)
        
        self.b1=Button(self.button_frame,text='Add',font=('aria',12,'bold'),bd=3,bg="grey",fg='black',width=11,height=3,command=add1)
        self.b1.grid(row=0,column=0,padx=4,pady=2)
        
        self.b2=Button(self.button_frame,text="Generate Bill",font=('aria',12,'bold'),bd=3,bg="grey",fg='black',width=11,height=3,command=gen)
        self.b2.grid(row=0,column=1,padx=4,pady=2)
        
        self.b3=Button(self.button_frame,text='Clear',font=('aria',12,'bold'),bd=3,bg="grey",fg='black',width=11,height=3,command=clr)
        self.b3.grid(row=0,column=2,padx=4,pady=2)
        
        self.b4=Button(self.button_frame,text='Total',font=('aria',12,'bold'),bd=3,bg="grey",fg='black',width=11,height=3,command=total)
        self.b4.grid(row=1,column=0,padx=4,pady=2)
        
        self.b5=Button(self.button_frame,text='Reset',font=('aria',12,'bold'),bd=3,bg="grey",fg='black',width=11,height=3,command=reset)
        self.b5.grid(row=1,column=1,padx=4,pady=2)
        
        self.b6=Button(self.button_frame,text='Save',font=('aria',12,'bold'),bd=3,bg="grey",fg='black',width=11,height=3,command=save)
        self.b6.grid(row=1,column=2,padx=4,pady=2)
        
        self.b1.config(state="disabled")
        self.b4.config(state="disabled")
        self.b6.config(state="disabled")
        
        #-----------------------------------Calculator---------------------------------------
        
        
        self.calc_frame=Frame(self.win,bd=8,bg="lightgrey",relief="groove")
        self.calc_frame.place(x=570,y=95,width=720,height=300)
        
        self.ent1=Entry(self.calc_frame,bd=15,bg="black",textvariable=calc_var,font=('aria',15,'bold'),width=45,fg="white",justify="right")
        self.ent1.grid(row=0,column=1,columnspan=3)
        
        def  btnclick(numbers):
             global operator
             operator=operator + str(numbers)
             calc_var.set(operator)

        def clrdisplay():
            global operator
            operator=""
            calc_var.set("")

        def eqals():
            global operator
            sumup=str(eval(operator))

            calc_var.set(sumup)
            operator = ""
        self.bb1=Button(self.calc_frame,bg="black",text="7",fg="white",bd=8,width=12,height=1,font=('aria',15,'bold'),command = lambda: btnclick(7))
        self.bb1.grid(row=1,column=0,padx=2,pady=2)
        #self.bb1.bind("<Button-1>",press_btn)
        self.bb2=Button(self.calc_frame,bg="black",text="8",fg="white",bd=8,width=12,height=1,font=('aria',15,'bold'),command = lambda: btnclick(8))
        self.bb2.grid(row=1,column=1,padx=2,pady=2)
        #self.bb2.bind("<Button-1>",press_btn)
        self.bb3=Button(self.calc_frame,bg="black",text="9",fg="white",bd=8,width=12,height=1,font=('aria',15,'bold'),command = lambda: btnclick(9))
        self.bb3.grid(row=1,column=2,padx=2,pady=2)
        #self.bb3.bind("<Button-1>",press_btn)
        self.bb4=Button(self.calc_frame,bg="black",text="+",fg="white",bd=8,width=12,height=1,font=('aria',15,'bold'),command = lambda: btnclick("+"))
        self.bb4.grid(row=1,column=3,padx=2,pady=2)
        #self.bb4.bind("<Button-1>",press_btn)
        
        self.bb5=Button(self.calc_frame,bg="black",text="4",fg="white",bd=8,width=12,height=1,font=('aria',15,'bold'),command = lambda: btnclick(4))
        self.bb5.grid(row=2,column=0,padx=2,pady=2)
        #self.bb5.bind("<Button-1>",press_btn)
        self.bb6=Button(self.calc_frame,bg="black",text="5",fg="white",bd=8,width=12,height=1,font=('aria',15,'bold'),command = lambda: btnclick(5))
        self.bb6.grid(row=2,column=1,padx=2,pady=2)
       # self.bb6.bind("<Button-1>",press_btn)
        self.bb7=Button(self.calc_frame,bg="black",text="6",fg="white",bd=8,width=12,height=1,font=('aria',15,'bold'),command = lambda: btnclick(6))
        self.bb7.grid(row=2,column=2,padx=2,pady=2)
        #self.bb7.bind("<Button-1>",press_btn)
        self.bb8=Button(self.calc_frame,bg="black",text="-",fg="white",bd=8,width=12,height=1,font=('aria',15,'bold'),command = lambda: btnclick("-"))
        self.bb8.grid(row=2,column=3,padx=2,pady=2)
       # self.bb8.bind("<Button-1>",press_btn)
        
        self.bb9=Button(self.calc_frame,bg="black",text="1",fg="white",bd=8,width=12,height=1,font=('aria',15,'bold'),command = lambda: btnclick(1))
        self.bb9.grid(row=3,column=0,padx=2,pady=2)
        #self.bb9.bind("<Button-1>",press_btn)
        self.bb10=Button(self.calc_frame,bg="black",text="2",fg="white",bd=8,width=12,height=1,font=('aria',15,'bold'),command = lambda: btnclick(2))
        self.bb10.grid(row=3,column=1,padx=2,pady=2)
        #self.bb10.bind("<Button-1>",press_btn)
        self.bb11=Button(self.calc_frame,bg="black",text="3",fg="white",bd=8,width=12,height=1,font=('aria',15,'bold'),command = lambda: btnclick(3))
        self.bb11.grid(row=3,column=2,padx=2,pady=2)
        #self.bb11.bind("<Button-1>",press_btn)
        self.bb12=Button(self.calc_frame,bg="black",text="*",fg="white",bd=8,width=12,height=1,font=('aria',15,'bold'),command = lambda: btnclick("*"))
        self.bb12.grid(row=3,column=3,padx=2,pady=2)
        #self.bb12.bind("<Button-1>",press_btn)
        
        self.bb13=Button(self.calc_frame,bg="black",text="0",fg="white",bd=8,width=12,height=1,font=('aria',15,'bold'),command = lambda: btnclick(0))
        self.bb13.grid(row=4,column=0,padx=2,pady=2)
        #self.bb13.bind("<Button-1>",press_btn)
        self.bb14=Button(self.calc_frame,bg="black",text=".",fg="white",bd=8,width=12,height=1,font=('aria',15,'bold'),command = lambda: btnclick("."))
        self.bb14.grid(row=4,column=1,padx=2,pady=2)
        #self.bb14.bind("<Button-1>",press_btn)
        self.bb15=Button(self.calc_frame,bg="black",text="C",fg="white",bd=8,width=12,height=1,font=('aria',15,'bold'),command = clrdisplay)
        self.bb15.grid(row=4,column=2,padx=2,pady=2)
        #self.bb15.bind("<Button-1>",press_btn)
        self.bb16=Button(self.calc_frame,bg="black",text="/",fg="white",bd=8,width=12,height=1,font=('aria',15,'bold'),command = lambda: btnclick("/"))
        self.bb16.grid(row=4,column=3,padx=2,pady=2)
        #self.bb16.bind("<Button-1>",press_btn)
        
        self.bb17=Button(self.calc_frame,bg="black",text="=",fg="white",bd=8,width=12,height=1,font=('aria',15,'bold'),command = eqals)
        self.bb17.grid(row=0,column=0,padx=2,pady=2)
        
        
        #-------------------BillFrame----------------------------------------------
        self.bill_frame=LabelFrame(self.win,text="Bill Area",font=('aria',18,"bold"),bd=8,bg="lightgrey",relief="groove")
        self.bill_frame.place(x=570,y=400,width=720,height=280)
        self.y_scroll= Scrollbar(self.bill_frame,orient="vertical")
        self.bill_txt = Text(self.bill_frame,bg="white",yscrollcommand=self.y_scroll.set)
        self.y_scroll.config(command=self.bill_txt.yview)
        self.y_scroll.pack(side=RIGHT,fill=Y)
        self.bill_txt.pack(fill=BOTH,expand=TRUE)
        default_bill()
if __name__ == "__main__":
    main()
```
